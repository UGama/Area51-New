```python
def fix_shifted_rows(df, date_column_name):
    df = df.copy()
    for idx, row in df.iterrows():
        date_val = str(row[date_column_name])
        if isinstance(date_val, str) and date_val.strip().startswith('$'):
            cols = df.columns.tolist()
            date_idx = cols.index(date_column_name)
            row_values = row.values.tolist()
            shifted = row_values[:date_idx] + [None] + row_values[date_idx:-1]
            df.loc[idx] = shifted
    return df
```

```python
from docx import Document
from docx.shared import Inches, Pt
from docx.enum.text import WD_PARAGRAPH_ALIGNMENT
from docx.oxml.ns import qn
from docx.oxml import OxmlElement

def export_df_to_word(df, filename, heading):
    df.columns = df.columns.str.replace('_', ' ').str.title()
    df = df.applymap(lambda x: '{:,.2f}'.format(x) if isinstance(x, float) and not x.is_integer()
                     else '{:,.0f}'.format(x) if isinstance(x, float)
                     else str(x))

    doc = Document()
    doc.add_heading(heading, 0)

    table = doc.add_table(rows=1, cols=len(df.columns))
    table.style = 'Table Grid'
    hdr_cells = table.rows[0].cells
    for i, column_name in enumerate(df.columns):
        hdr_cells[i].text = str(column_name)
        para = hdr_cells[i].paragraphs[0]
        para.alignment = WD_PARAGRAPH_ALIGNMENT.CENTER
        run = para.runs[0]
        run.bold = True
        run.font.size = Pt(10)
    for _, row in df.iterrows():
        row_cells = table.add_row().cells
        for i, val in enumerate(row):
            cell = row_cells[i]
            cell.text = str(val)
            cell.paragraphs[0].alignment = WD_PARAGRAPH_ALIGNMENT.CENTER

    doc.save(filename)
```

```python
import tabula
import pandas as pd

header_pages = [232, 272, 365]
pages = list(range(232, 382))
tables = []
current_header = None

for i in pages:
    try:
        # print(f"Processing page {i}...")
        df = tabula.read_pdf("Venue Forecasts and Labour.pdf", pages=i, multiple_tables=False)[0]
        if 'key' in df.columns:
            df = df.drop(columns='key')

        if i in header_pages:
            if df.iloc[0].equals(pd.Series(df.columns)): # no colunms
                current_header = df.iloc[0].tolist()
                df = df[1:].reset_index(drop=True)
            else: # have colums
                current_header = [str(c).strip() for c in df.columns]
                # print("✅Table header detected automatically")

            current_header[0] = "Location"
            df.columns = current_header
        elif current_header is not None:
            df.columns = current_header
        else:
            print(f"⚠️ Page {i} has no header and no previous header. Skipping.")
            continue

        # print(current_header)
        df = df.iloc[:, :5]
        df = df.reset_index(drop=True)
        df.columns = [str(col).strip() for col in df.columns]
        df.columns = df.columns.str.replace('_', ' ')
        if 'Bud Lab Cost' in df.columns:
            df = df.rename(columns={'Bud Lab Cost': 'Bud Lab Cost Mon-Sat'})
        if 'Unnamed: 4' in df.columns:
            df = df.drop(columns=['Unnamed: 4'])
        tables.append(df)

    except Exception as e:
        print(f"❌ Failed on page {i}: {e}")

labour_cost_bud = pd.concat(tables, ignore_index=True)
labour_cost_bud
```

```python
pages = [23, 26, 28, 31, 34, 37, 40, 43, 46, 48, 50, 53, 56, 59, 62, 64, 67, 70, 73, 76, 79, 82]

venue_datalist = []

for i in pages:
    venue = tabula.read_pdf("Venue Forecasts and Labour.pdf", pages = i, multiple_tables=False)[0]
    
    venue.columns = venue.iloc[0].str.replace(r'\r', ' ', regex=True)
    venue = venue.iloc[1:].reset_index(drop=True)
    venue.columns = venue.columns.str.strip()

    venue_name = venue.iloc[0, 0]

    venue_Jun = venue.head(30).copy()
    
    if 'Actual Labour' not in venue_Jun.columns and 'Actual Labour (From ENTO)' in venue_Jun.columns:
        venue_Jun['Actual Labour'] = venue_Jun['Actual Labour (From ENTO)']

    venue_Jun = fix_shifted_rows(venue_Jun, date_column_name='Date Entered')
    venue_Jun.loc[:, 'Actual Labour'] = venue_Jun['Actual Labour'].replace(r'[\$,]', '', regex=True)
    venue_Jun.loc[:, 'Actual Labour'] = pd.to_numeric(venue_Jun['Actual Labour'], errors='coerce')

    venue_Jun.loc[:, 'Actual Sales'] = venue_Jun['Actual Sales'].replace(r'[\$,]', '', regex=True)
    venue_Jun.loc[:, 'Actual Sales'] = pd.to_numeric(venue_Jun['Actual Sales'], errors='coerce')

    max_lval = venue_Jun['Actual Labour'].max()
    min_lval = venue_Jun['Actual Labour'].min()
    mean_lval = venue_Jun['Actual Labour'].mean()
    sum_lval = venue_Jun['Actual Labour'].sum()

    max_sval = venue_Jun['Actual Sales'].max()
    min_sval = venue_Jun['Actual Sales'].min()
    mean_sval = venue_Jun['Actual Sales'].mean()
    sum_sval = venue_Jun['Actual Sales'].sum()

    venue_Jun["Labour/Sales Ratio"] = venue_Jun["Actual Labour"] / venue_Jun["Actual Sales"]
    
    max_ratio = venue_Jun["Labour/Sales Ratio"].max()
    min_ratio = venue_Jun["Labour/Sales Ratio"].min()
    if sum_sval == 0:
        total_ratio = 0
    else:
        total_ratio = sum_lval / sum_sval

    if venue_Jun['Actual Labour'].dropna().empty:
        max_ldate = 'no data'
        min_ldate = 'no data'
    else:
        max_ldate = venue_Jun.loc[venue_Jun['Actual Labour'].idxmax(), 'Date']
        min_ldate = venue_Jun.loc[venue_Jun['Actual Labour'].idxmin(), 'Date']

    if venue_Jun['Actual Sales'].dropna().empty:
        max_sdate = 'no data'
        min_sdate = 'no data'
    else:
        max_sdate = venue_Jun.loc[venue_Jun['Actual Sales'].idxmax(), 'Date']
        min_sdate = venue_Jun.loc[venue_Jun['Actual Sales'].idxmin(), 'Date']

    if venue_Jun["Labour/Sales Ratio"].dropna().empty:
        max_rdate = 'no data'
        min_rdate = 'no data'
        count_ratio_gt1 = 'no data'
    else:
        max_rdate = venue_Jun.loc[venue_Jun["Labour/Sales Ratio"].idxmax(), 'Date']
        min_rdate = venue_Jun.loc[venue_Jun["Labour/Sales Ratio"].idxmin(), 'Date']
        count_ratio_gt1 = (venue_Jun["Labour/Sales Ratio"] > 1).sum()

    labour_days = venue_Jun['Actual Labour'].dropna().shape[0]
    sales_days = venue_Jun['Actual Sales'].dropna().shape[0]

    column_names = ['Parties Booked (not weather dependent)', 'Programs Booked (not weather dependent)',
                   'Schools & Groups Booked (not weather dependent)']

    BookList = []
    for column_name in column_names:
        venue_Jun.loc[:, column_name] = venue_Jun[column_name].replace(r'[\$,]', '', regex=True)
        venue_Jun.loc[:, column_name] = pd.to_numeric(venue_Jun[column_name], errors='coerce')
        valid_data = venue_Jun[column_name][(venue_Jun[column_name].notna()) & (venue_Jun[column_name] != 0)]
        valid_days = valid_data.shape[0]
        BookList.append(valid_days)
        
        if valid_days == 0:
            BookList.append(0)
            BookList.append(0)
        else:
            BookList.append(valid_data.sum())
            BookList.append(valid_data.max())
    
    # save results
    venue = pd.DataFrame({
        'Location': [venue_name],
        'Collected_Labour_Days': [labour_days],
        'Max_Labour_Cost': [max_lval],
        'Max_Labour_Date': [max_ldate],
        'Min_Labour_Cost': [min_lval],
        'Min_Labour_Date': [min_ldate],
        'Mean_Labour_Cost': [mean_lval],
        'Sum_Labour_Cost': [sum_lval],
        
        'Collected_Sales_Days': [sales_days],
        'Max_Sales': [max_sval],
        'Max_Sales_Date': [max_sdate],
        'Min_Sales': [min_sval],
        'Min_Sales_Date': [min_sdate],
        'Mean_Sales': [mean_sval],
        'Sum_Sales': [sum_sval],

        'Parties_Booked_Days': BookList[0],
        'Sum_Parties_Booked': BookList[1],
        'Max_Parties_Booked': BookList[2],

        'Programs_Booked_Days': BookList[3],
        'Sum_Programs_Booked': BookList[4],
        'Max_Programs_Booked': BookList[5],

        'Schools&Groups_Booked_Days': BookList[6],
        'Sum_Schools&Groups_Booked': BookList[7],
        'Max_Schools&Groups_Booked': BookList[8],

        'Max_L/S': [max_ratio],
        'Max_L/S_Date': [max_rdate],
        'Min_L/S': [min_ratio],
        'Min_L/S_Date': [min_rdate],
        'total_ratio': [total_ratio],
        'L/S_gt1': [count_ratio_gt1],
    })
    venue_datalist.append(venue)


venue_summary = pd.concat(venue_datalist, ignore_index=True)
venue_summary = venue_summary.sort_values(by='Mean_Sales', ascending=False)
venue_summary.reset_index(drop=True, inplace=True)

venue_summary
```
