```python
school_holiday_ranges = [
    # 2022
    ('2022-06-25', '2022-07-10'),
    ('2022-09-17', '2022-10-03'),
    ('2022-12-10', '2023-01-22'),
    # 2023
    ('2023-04-01', '2023-04-16'),
    ('2023-06-24', '2023-07-09'),
    ('2023-09-16', '2023-10-02'),
    ('2023-12-09', '2024-01-21'),
    # 2024
    ('2024-03-29', '2024-04-14'),
    ('2024-06-22', '2024-07-07'),
    ('2024-09-14', '2024-09-29'),
    ('2024-12-14', '2025-01-27'),
    # 2025
    ('2025-04-05', '2025-04-21'),
    ('2025-06-28', '2025-07-13'),
]

holiday_dates = set()
for start, end in school_holiday_ranges:
    dates = pd.date_range(start=start, end=end)
    holiday_dates.update(dates)

df_grouped['is_school_holiday'] = df_grouped['date'].isin(holiday_dates).astype(int)
df_grouped[:-1]
```

```python
df_merged1 = pd.merge(df_grouped, df_maxtemp, on='date', how='left')
df_merged1.head(10)
```

```python
print(df_merged1.isna().sum())
```

```python
import numpy as np

temps = df_merged1['max_temp'].values.copy()

for i in range(1, len(temps) - 1):
    if np.isnan(temps[i]) and not np.isnan(temps[i - 1]) and not np.isnan(temps[i + 1]):
        temps[i] = (temps[i - 1] + temps[i + 1]) / 2
        
df_merged1['max_temp'] = temps
print(df_merged1.isna().sum())
```
